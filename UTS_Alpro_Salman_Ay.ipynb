```python
# Meminta input usia dari pengguna
usia = int(input("Masukkan usia Anda: "))

# Menentukan harga tiket berdasarkan usia
if usia < 12:
    harga_tiket = 25000  # 25k untuk Anak-anak
    predikat = "Anak-anak"
elif usia >= 12 and usia <= 17:
    harga_tiket = 40000  # 40k untuk remaja
    predikat = "Remaja"
elif usia >= 18 and usia <= 59:
    harga_tiket = 60000  # 60k untuk dewasa
    predikat = "Dewasa"
else:
    harga_tiket = 0  # Gratis untuk lansia
    predikat = "Lansia"

# Menampilkan harga tiket
print(f"Usia Anda: {usia} tahun ({predikat}) ")
print(f"Harga tiket yang harus dibayar: Rp {harga_tiket:,}")
```
